```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([about])   # in this we send a multiple text in a tokenizer so we pass in form of list
```

```python
# we will find out sentences present in the dataset
input_sentences = []

for sentence in about.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]                    # this word covert into number and [0] is used to slice or come out single index

  for i in range(1, len(tokenized_sentence)):                                         # nested loop for picking up single line and then apply in this form [1,2,3]--->[1,2][1,2,3]  For this loop runs two times
    input_sentences.append(tokenized_sentence[:i+1])
```

```python
input_sentences
```

```python
# in above output the shape is not equal for this we apply "padding"  --> for padding we require max sentence length
max_len = max(len(x) for x in input_sentences)
print(max_len)
```

```python
# padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sentences, max_len, padding = 'pre')                                                                       # In this we pass 3 parameter in the pad sequence
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

Convert the data into **Supervised data** in with X and y

```python
y = padded_input_sequences[:,-1]
```

Text_generator --> supervised--> classification ---> multiclass_classification   "So we use **OneHot Encoding"

Convert output into categorical which can help to predict

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes = 267)               # num_classes = size of the (tokenizer +1) --> tokenizer.word_index = index start from 1 n
```

```python
y.shape
```

```python
y
```

```python
# building the architecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(267, 100, input_length = 56))      # 267 total no. of word counts , output dense vector 100 number  input_length =56 becoz each row contain 56 words
model.add(LSTM(150))                                  # her lstm gate per kitne node honge
#model.add(LSTM(150))
model.add(Dense(267, activation = 'softmax'))        # 267 utne node dalne honge jitne apne vocabulary words h
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X,y,epochs=100)
```
