* Function Api with **"Multiple input and single output"**
* In this case we contain the multiple input branch and single output value 

```python
from keras.layers import *
from keras.models import Model
```

```python
# define two set of input

inputA = Input(shape = (32,))
inputB = Input(shape = (128,))
```

```python
# from this we build the ------> no.of branch = no. of input 
# first branch operate on the first input 

x = Dense(8,activation = 'relu')(inputA)
x1 = Dense(4, activation ='relu')(x)
```

```python
# the second branch operates on the second input 

y = Dense(64,  activation = 'relu')(inputB)
y1 = Dense(32, activation ='relu')(y)
y2 = Dense(4, activation = 'relu')(y1)
```

```python
# combine the output of the two branch 

combined = concatenate([x1,y2])
```

```python
# apply a Fully connected layer and then regression prediction on the Combined output

z = Dense(2, activation ='relu')(combined)
z1 = Dense(1, activation = 'linear')(z)
```

```python
# our model will accept the input of the two branches and then output a single value (linear)
model = Model(inputs = [inputA,inputB], outputs = z1)
```

```python
# Plot

from keras.utils import plot_model
plot_model(model, show_shapes = True)
```
