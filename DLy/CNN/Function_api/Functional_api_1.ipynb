* Functional Api :-- basically it is "non-linear" topology in which no. of input and ouput is not fix.

* Sequential : - for "linear " Data we we use the Sequential training model.

```python
from keras.models import Model
```

```python
from keras.layers import *      # import all the layer 
```

```python
x = Input(shape=(3,))           # it mean input shape contain 3 col
```

```python
hidden1 = Dense(128, activation = 'relu')(x)      # 1st layer with 128 neuron and activaiton - relu and this layer is connected to x 
hidden2 = Dense(64, activation = 'relu')(hidden1)   # 2nd layer with 64 neuron anad activaiton relu and this layer is connected to hidden1

output1 = Dense(1,activation = 'linear')(hidden2)
output2 = Dense(1,activation = 'sigmoid')(hidden2)

# this contain 2 output 1 st one is linear regression (age) and 2nd for the (place)--sigmoid classificattion data and both output connected to the 2 hidden layer
```

```python
model = Model(inputs = x ,outputs = [output1,output2])
# it say it contain 1 input x and 2 output 
```

```python
model.summary()
```

```python
# this is keras.utils to visulaize what we have made (plot_model)

from keras.utils import plot_model
plot_model(model)
```

```python
plot_model(model, show_shapes = True)
```
