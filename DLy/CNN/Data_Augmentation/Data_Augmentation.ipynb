````python
# this is not in "numpy" format it pil type 
datagen = ImageDataGenerator(
    rotation_range = 30,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
   # vertical_flip = True,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
  # fill_mode = 'nearest'/'reflect'/'constant'
)

# rotation range -- angle by which you rotate
# shear imge -- distord the image 1 scale per 0.2 la distord image
# zoom range -- zoom in or zoom out
# vertical flip -- not possible because cat ulti nahi hoti 
# fill_mode = 'nearest'/'reflect'/'constant'  chooice any one
)
````

# convert the image type

````python
img = image.img_to_array(img)
# convert the image into array type
````

````python
img.shape
````

````python
input_batch = img.reshape(1,200,200,3)
# 1 -- it mean no.of image we paas
````

datagen has object **"flow" and "flow_from_directory"**
* "flow" is generaly use when we work on the single image  or directly on numpy
* "flow from directory" used when we work on the directory

> Indented block



````python
i = 0

for output in datagen.flow(input_batch, batch_size = 1, save_to_dir = '/content/'):

  i = i+1

  if i == 10:
    break
````

** Data Argumentation for the Entire Directory **

````python
# this code is not runable use this as reference

batch_size = 16

train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)

# test_generator -- in this we not doing any type of change 
test_datagen = ImageDataGenerator(rescale = 1./255)

train_generator = train_datagen.flow_from_directory(
    'data/train',    #this is the target directory (path)
    target_size = (150,150),   # all images will be resized to 150*150
    batch_size = batch_size,
    class_mode = 'binary'    # since we use "binary_crossentropy loss", we need binary labels
)

# this is a similar generator, for validation data

validation_generator = test_datagen.flow_from_directory(
    
    'data/valid',
    target_size = (150,150)
    batch_size = batch_size,
    class_mode = 'binary'
)
````
